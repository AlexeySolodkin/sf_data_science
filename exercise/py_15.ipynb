```python
class DepartmentReport():
    def __init__(self, company_name):
        self.revenues = []
        self.company_name = company_name
    def add_revenue(self, amount):
        self.revenues.append(amount)
    def average_revenue(self):
        result = int(round(sum(self.revenues) / len(self.revenues), 0))
        return f'Average department revenue for {self.company_name}: {result}'
```

```python
report = DepartmentReport("Danon")
report.add_revenue(1_000_000)
report.add_revenue(400_000)

print(report.average_revenue())
```

```python
class User():
    def __init__(self, email, password, balance):
        self.email = email
        self.password = password
        self.balance = balance
    def login(self, email, password):
        return self.email == email and self.password == password
    def update_balance(self, amount):
        self.balance += amount
        return self.balance
```

```python
class IntDataFrame():
    def __init__(self, column):
        self.column = column
        self.to_int()
    def to_int(self):
        self.column = [int(value) for value in self.column]
    def count(self):
        number_not_null = 0
        for value in self.column:
            if value != 0:
                number_not_null += 1
        return number_not_null
    def unique(self):
        return len(set(self.column))
```

```python
df = IntDataFrame([4.7, 4, 3, 0, 2.4, 0.3, 4])

df.count()
```

```python
class OwnLogger():
    def __init__(self):
        self.mesages_dict = {'info': [], 'warning': [], 'error': []}
        self.last_mesage = None
    def log(self, message, level):
        self.mesages_dict[level].append(message)
        self.last_mesage = message
    def show_last(self, level='all'):
        if level == 'all':
            return self.last_mesage
        elif len(self.mesages_dict[level]) == 0:
            return None
        else:
            return self.mesages_dict[level][-1]
```

```python
logger = OwnLogger()
logger.log("System started", "info")
logger.show_last("error")
```

```python
logger.log("Connection instable", "warning")
logger.log("Connection lost", "error")
```

```python
logger.show_last()
```

```python
def walk_desc(path=None):
    start_path = path if path is not None else os.getcwd()

    for root, dirs, files in os.walk(start_path):
        print("Текущая директория", root)
        print("---")

        if dirs:
            print("Список папок", dirs)
        else:
            print("Папок нет")
        print("---")

        if files:
            print("Список файлов", files)
        else:
            print("Файлов нет")
        print("---")

        if files and dirs:
            print("Все пути:")
        for f in files:
            print("Файл ", os.path.join(root, f))
        for d in dirs:
            print("Папка ", os.path.join(root, d))
        print("===")

walk_desc()
```
