```python
weekend_list = [5, 6]
citibike_df['weekend'] = citibike_df['starttime'].dt.dayofweek.apply(lambda x: 1 if x in weekend_list else 0)
citibike_df[citibike_df['weekend'] == 1].shape[0]
```

```python
def time_of_day(hour):
    night_list = list(range(0, 7))
    morning_list = list(range(7, 13))
    day_list = list(range(13, 19))
    evening_list = list(range(19, 24))
    if hour in night_list:
        return 'night'
    if hour in morning_list:
        return 'morning'
    if hour in day_list:
        return 'day'
    if hour in evening_list:
        return 'evening'
```

```python
citibike_df['time_of_day'] = citibike_df['starttime'].dt.hour.apply(time_of_day)
trip_day = citibike_df[citibike_df['time_of_day'] == 'day'].shape[0]
trip_night = citibike_df[citibike_df['time_of_day'] == 'night'].shape[0]
round(trip_day / trip_night, 0)
```
