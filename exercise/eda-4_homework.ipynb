```python
# Загружаем библиотеки
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, pearsonr, ttest_ind
```

```python
petersburg = [0.0974, 0.1352, 0.0817, 0.1016, 0.0968, 0.1064, 0.105]
magadan = [0.1033, 0.0915, 0.0781, 0.0685, 0.0677, 0.0697, 0.0764, 0.0689]
```

```python
print(f'Размерность списка petersburg составляет {len(petersburg)} элементов.')
print(f'Размерность списка magadan составляет {len(magadan)} элементов.')
```

```python
# Размерность списка petersburg меньше, поэтому добавляем в него среднее значение
petersburg.append(np.mean(petersburg))
```

```python
# Создаем DataFrame
data = pd.DataFrame({'petersburg': petersburg, 'magadan': magadan})
```

```python
data
```

```python
# проверяем данные на нормальность, для этого будем применять Тест Шапиро-Уилка, 
# так как наша выборка имеет небольшой размер.
H0 = 'Данные распределены нормально'
Ha = 'Данные не распределены нормально'
alpha = 0.05

s, p = shapiro(data)

if p > alpha:
    print(f'p = {p:.3f}. {H0}')
else:
    print(f'p = {p:.3f}. {Ha}')
```

Так как мы имеем непрерывные нормально распределенные признаки, для определения наличия связи между признаками будем использовать коэффициент корреляции Пирсона.

```python
# Постановка гипотез
H0 = 'Осутсвует зависимость между переменными'
Ha = 'Не осутсвует зависимость между переменными'

cor, p = pearsonr(data['petersburg'], data['magadan'])

print(f'Коэффициент корреляции Пирсона = {cor:.3f}, p-value = {p:.3f}')
```

```python
if p>alpha:
    print(f"{p:.3f} > {alpha}. Мы не можем отвергнуть нулевую гипотезу об отсутсвии зависимости между переменными.")
else:
    print(f"{p:.3f} <= {alpha}. Мы отвергаем нулевую гипотезу об отсутсвии зависимости между переменными.")
```

Проведем сравнение выборок.

Так как наши выборки принадлежат к одной совокупности, имеют нормальный закон распределения и объем выборок не превышает 30 наблюдений, для сравнения выборок будем использовать независимый t-тест.

```python
# Постановка гипотез
H0 = 'Нет значимой разницы между средним размером раковины мидий в Петербурге и Магадане.'
Ha = 'Есть значимая разница между средним размером раковины мидий в Петербурге и Магадане.'

results = ttest_ind(data['petersburg'], data['magadan'], equal_var=True)

p = round(results[1], 3)

if p>alpha:
    print(f'{p:.3f} > {alpha}. Мы не можем отвергнуть нулевую гипотезу об отсутсвии значимой разницы между средним размером раковин мидий.')
else:
    print(f'{p:.3f} <= {alpha}. Мы отвергаем нулевую гипотезу об отсутсвии значимой разницы между средним размером раковин мидий.')
```
