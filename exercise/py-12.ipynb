```python
import pandas as pd
```

```python
melb_data = pd.read_csv('data/melb_data_fe.csv', sep=',')
```

```python
melb_df = melb_data.copy()
melb_df.head()
```

```python
melb_df.info()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
```

```python
melb_df['Date'].dt.quarter.value_counts()
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # список столбцов, которые мы не берём во внимание
max_unique_count = 150 # задаём максимальное число уникальных категорий
for col in melb_df.columns: # цикл по именам столбцов
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: # проверяем условие
        melb_df[col] = melb_df[col].astype('category') # преобразуем тип столбца
melb_df.info()
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
building_area_1558 = round(
    melb_df.sort_values(
        by='AreaRatio',
        ascending=False,
        ignore_index=True
    )
    .loc[1158, 'BuildingArea'],
    0
)
building_area_1558
```

```python
mask_type = melb_df['Type'] == 'townhouse'
mask_rooms = melb_df['Rooms'] > 2
melb_df[mask_type & mask_rooms].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True
).loc[18, :]
```

```python
melb_df.groupby('Rooms')['Price'].mean()
```

```python
melb_df.groupby('Regionname')['Lattitude'].std()
```

```python
mask_date = (melb_df['Date'] >= '2017-05-01') & (melb_df['Date'] <= '2017-09-01') 
melb_df[mask_date].groupby('SellerG')['Price'].sum().sort_values(ascending=True)
```

```python
melb_df.pivot_table(
    values='BuildingArea',
    index='Rooms',
    columns='Type',
    aggfunc='median',
    fill_value=0
)
```

```python
melb_df.pivot_table(
    values='Price',
    index='SellerG',
    columns='Type',
    aggfunc='median',
    fill_value=0
).sort_values(by='unit', ascending=False)
```
